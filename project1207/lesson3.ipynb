```python

import requests
url="https://www.ytower.com.tw/recipe/iframe-recipe.asp?seq=A03-1644"
response=requests.get(url)
response.encoding='big5' #編碼轉成big5
if response.ok:
    html_code=response.text #記得要把抓下來的文字轉成str
    print(type(html_code))
else:
    print("下載失敗")
```

```python
from bs4 import BeautifulSoup
#找宮保雞丁
soup=BeautifulSoup(html_code,'html.parser')
recipe_groups=soup.find(name="div",id="recipe_name")
recipe_groups.get_text()
```

```python
#name 標籤名稱
ingredients=soup.findAll(name="ul",class_="ingredient")
# print(ingredient)
for ingredient in ingredients:
    ingredient_list=[]
    for li in ingredient.findAll(name="li"):
        ingredient_list.append(li.text.strip())
        # print(type(li.text)) #<class 'str'>
        print(li.text.strip()) #strip() 是str內建的方法 可以把多餘的空白去掉

    print(ingredient_list)
```
